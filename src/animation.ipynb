```python
x = np.linspace(0, 2*np.pi, 100)
fig = plt.figure()
lines = plt.plot([])
line = lines[0]

plt.axis("scaled")
plt.xlim(0, 4)
plt.ylim(-1.5, 1.5)

def animate(frame):
    # update plot
    # I can plot robot's position based on the csv data with the animation
    # I can 
    y = np.sin(x + 2*np.pi * frame/100)
    line.set_data((x,y))
    
anim = FuncAnimation(fig, animate, frames=100, interval=20)
video = anim.to_html5_video()
html = display.HTML(video)
display.display(html)
plt.close()
```
