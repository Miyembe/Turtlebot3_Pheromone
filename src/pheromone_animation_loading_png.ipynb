
# Animated image using a precomputed list of images



```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
import os
from IPython import display

# Importing settings
time_str = '20210412-131508'
com_path = "/home/sub/catkin_ws/src/Turtlebot3_Pheromone/src/plot_script"
dir_path = os.path.join(com_path, time_str)
num_files = len([name for name in os.listdir(dir_path) if os.path.isfile(os.path.join(dir_path, name))])
```

```python
# Figure settings
fig, ax = plt.subplots()

# Remove axis
plt.axis('off')

plt.xticks([])
    
```

```python
# Make the images an animation
ims = []
for i in range(1, num_files):
        image = plt.imread(dir_path +'/'+ time_str + "_{}.png".format(i))
        im = ax.imshow(image, animated=True)
        ims.append([im])
print(len(ims))
ani = animation.ArtistAnimation(fig, ims, interval=80, blit=True,
                                repeat_delay=1000)
plt.show()
```

```python
video = ani.to_html5_video()
html = display.HTML(video)
display.display(html)
plt.close()
```
